### Step 3 : Converting Sentence into Embeddings
- File : glove.6B.50d.txt
- 6 billion words
- Each word is a 50 dimensional vector

```python
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove.6B.zip
```

```python
embeddings_index = {}

f = open('glove.6B.50d.txt', encoding='utf-8')

for line in f:
    
    values = line.split()
    
    word = values[0]
    
    word_vector = np.asarray(values[1:], dtype='float32')
    
    embeddings_index[word] = word_vector

f.close()
```

```python
embeddings_index["food"].shape
```

### Step 4 : Converting sentences into vectors (Embedding Layer Output)

```python
def embedding_output(X):
    
    maxLen = 10 # As our dataset is very small
    emb_dim = 50 # size of each word vector (Glove Vector)
    embedding_out = np.zeros((X.shape[0], maxLen, emb_dim))
    
    for i in range(X.shape[0]):
        # for each sentence
        sent = X[i].split()
        
        for j in range(maxLen):
            # for each word of the sentence till maxLen
            
            try:
                embedding_out[i][j] = embeddings_index[sent[i][j].lower()]
            except:
                embedding_out[i][j] = np.zeros((50,))
                
    return embedding_out
```

```python
embeddings_matrix_train = embedding_output(X_train.values)
embeddings_matrix_test = embedding_output(X_test.values)
```

```python
print(embeddings_matrix_train.shape)
print(embeddings_matrix_test.shape)
```

### Step 5 : Define the RNN/LSTM Model

```python
# Y should be one hot vector
from keras.utils import to_categorical

Y_train = to_categorical(Y_train, num_classes=5)
Y_test = to_categorical(Y_test, num_classes=5)

print(Y_train.shape)
print(Y_test.shape)
```

```python
from keras.layers import LSTM, Dropout, Dense, Activation
from keras.models import Sequential
```

```python
model = Sequential()

# Stacked LSTM
model.add(LSTM(64,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(64))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.summary()
```

```python
from keras.optimizers import Adam
model.compile(optimizer=Adam(lr=0.01,),loss='categorical_crossentropy',metrics=['acc'])
```

```python
# from keras.callbacks import EarlyStopping, ModelCheckpoint
# earlyStopping = EarlyStopping(monitor= 'val_accuracy', patience=20)
# checkpoint = ModelCheckpoint("best_model.h5", monitor= 'val_loss', verbose=1, save_best_only= True)

## Training the model
model.fit(embeddings_matrix_train,Y_train,batch_size=32,epochs=100,shuffle=True,validation_split=0.1)
```

```python
model.evaluate(embeddings_matrix_test, Y_test)
```
